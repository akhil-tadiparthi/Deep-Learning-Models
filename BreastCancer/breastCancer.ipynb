```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo 
```

```python
# fetch dataset 
ionosphere = fetch_ucirepo(id=52) 
  
# data (as pandas dataframes) 
X = ionosphere.data.features 
y = ionosphere.data.targets 

# Encode the target variable
le = LabelEncoder()
y_binary = le.fit_transform(y)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.25, random_state=1)

# Reshape y_train and y_test to be one-dimensional
y_train = y_train.ravel()
y_test = y_test.ravel()

# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Different number of hidden layers
mlp1 = MLPClassifier(hidden_layer_sizes=(25, 20), activation='relu', max_iter=1000, random_state=1)
mlp1.fit(X_train, y_train)
y_pred1 = mlp1.predict(X_test)
mat1 = metrics.confusion_matrix(y_test, y_pred1)
precision1 = metrics.precision_score(y_test, y_pred1)
recall1 = metrics.recall_score(y_test, y_pred1)
accuracy1 = accuracy_score(y_test, y_pred1)
results1 = {
    "Confusion Matrix": mat1.tolist(),
    "Precision": precision1,
    "Recall": recall1,
    "Accuracy": accuracy1
}

mlp2 = MLPClassifier(hidden_layer_sizes=(50, 20), activation='relu', max_iter=1000, random_state=1)
mlp2.fit(X_train, y_train)
y_pred2 = mlp2.predict(X_test)
mat2 = metrics.confusion_matrix(y_test, y_pred2)
precision2 = metrics.precision_score(y_test, y_pred2)
recall2 = metrics.recall_score(y_test, y_pred2)
accuracy2 = accuracy_score(y_test, y_pred2)
results2 = {
    "Confusion Matrix": mat2.tolist(),
    "Precision": precision2,
    "Recall": recall2,
    "Accuracy": accuracy2

}

# Different number of neurons per hidden layer
mlp3 = MLPClassifier(hidden_layer_sizes=(25, 40), activation='relu', max_iter=1000, random_state=1)
mlp3.fit(X_train, y_train)
y_pred3 = mlp3.predict(X_test)
mat3 = metrics.confusion_matrix(y_test, y_pred3)
precision3 = metrics.precision_score(y_test, y_pred3)
recall3 = metrics.recall_score(y_test, y_pred3)
accuracy3 = accuracy_score(y_test, y_pred3)
results3 = {
    "Confusion Matrix": mat3.tolist(),
    "Precision": precision3,
    "Recall": recall3,
    "Accuracy": accuracy3

}

mlp4 = MLPClassifier(hidden_layer_sizes=(50, 40), activation='relu', max_iter=1000, random_state=1)
mlp4.fit(X_train, y_train)
y_pred4 = mlp4.predict(X_test)
mat4 = metrics.confusion_matrix(y_test, y_pred4)
precision4 = metrics.precision_score(y_test, y_pred4)
recall4 = metrics.recall_score(y_test, y_pred4)
accuracy4 = accuracy_score(y_test, y_pred4)
results4 = {
    "Confusion Matrix": mat4.tolist(),
    "Precision": precision4,
    "Recall": recall4,
    "Accuracy": accuracy4
}

# Same as above with different activation functions
mlp5 = MLPClassifier(hidden_layer_sizes=(25, 20), activation='tanh', max_iter=1000, random_state=1)
mlp5.fit(X_train, y_train)
y_pred5 = mlp5.predict(X_test)
mat5 = metrics.confusion_matrix(y_test, y_pred5)
precision5 = metrics.precision_score(y_test, y_pred5)
recall5 = metrics.recall_score(y_test, y_pred5)
accuracy5 = accuracy_score(y_test, y_pred5)
results5 = {
    "Confusion Matrix": mat5.tolist(),
    "Precision": precision5,
    "Recall": recall5,
    "Accuracy": accuracy5
}

mlp6 = MLPClassifier(hidden_layer_sizes=(50, 20), activation='tanh', max_iter=1000, random_state=1)
mlp6.fit(X_train, y_train)
y_pred6 = mlp6.predict(X_test)
mat6 = metrics.confusion_matrix(y_test, y_pred6)
precision6 = metrics.precision_score(y_test, y_pred6)
recall6 = metrics.recall_score(y_test, y_pred6)
accuracy6 = accuracy_score(y_test, y_pred6)
results6 = {
    "Confusion Matrix": mat6.tolist(),
    "Precision": precision6,
    "Recall": recall6,
    "Accuracy": accuracy6
}

mlp7 = MLPClassifier(hidden_layer_sizes=(25, 40), activation='tanh', max_iter=1000, random_state=1)
mlp7.fit(X_train, y_train)
y_pred7 = mlp7.predict(X_test)
mat7 = metrics.confusion_matrix(y_test, y_pred7)
precision7 = metrics.precision_score(y_test, y_pred7)
recall7 = metrics.recall_score(y_test, y_pred7)
accuracy7 = accuracy_score(y_test, y_pred7)
results7 = {
    "Confusion Matrix": mat7.tolist(),
    "Precision": precision7,
    "Recall": recall7,
    "Accuracy": accuracy7
}

mlp8 = MLPClassifier(hidden_layer_sizes=(50, 40), activation='tanh', max_iter=1000, random_state=1)
mlp8.fit(X_train, y_train)
y_pred8 = mlp8.predict(X_test)
mat8 = metrics.confusion_matrix(y_test, y_pred8)
precision8 = metrics.precision_score(y_test, y_pred8)
recall8 = metrics.recall_score(y_test, y_pred8)
accuracy8 = accuracy_score(y_test, y_pred8)
results8 = {
    "Confusion Matrix": mat8.tolist(),
    "Precision": precision8,
    "Recall": recall8,
    "Accuracy": accuracy8
}

# Confusion matrix visualizations
confusion_matrices = [mat1, mat2, mat3, mat4, mat5, mat6, mat7, mat8]
model_names = ["Model 1", "Model 2", "Model 3", "Model 4", "Model 5", "Model 6", "Model 7", "Model 8"]

for i in range(8):
    sns.heatmap(confusion_matrices[i].T, square=True, annot=True, fmt='d', cbar=False)
    plt.xlabel("True Label")
    plt.ylabel("Predicted Label")
    plt.title(f"Confusion Matrix - {model_names[i]}")
    plt.show()

results_list = [results1, results2, results3, results4, results5, results6, results7, results8]

for idx, result in enumerate(results_list):
    print(f"Model {idx + 1} Results:")
    print("Confusion Matrix:")
    print(result["Confusion Matrix"])
    print("Precision:", result["Precision"])
    print("Recall:", result["Recall"])
    print("Accuracy:", result["Accuracy"])
    print("\n")
```

```python
from sklearn.datasets import load_digits

# Digits Dataset - multi class classification
digits = load_digits()
X, y = digits.data, digits.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

train_sizes = [0.25, 0.5, 0.75, 1.0]
train_scores = []


plt.figure(figsize=(12, 8))

# Loop over the different sizes of the training set
for i, size in enumerate(train_sizes):
    X_train_partial = X_train[:int(size * X_train.shape[0])]
    y_train_partial = y_train[:int(size * len(y_train))]
    
    mlp = MLPClassifier(hidden_layer_sizes=(100,), activation='relu', batch_size=200, solver='adam', max_iter=200, random_state=1)
    
    loss_values = []
    for _ in range(200): # 200 epochs
        mlp.partial_fit(X_train_partial, y_train_partial, classes=np.unique(y))
        loss_values.append(mlp.loss_)
    
    plt.plot(loss_values, label=f"Training with {int(size * 100)}% of data")
    train_scores.append(mlp.score(X_test, y_test))

# plot
plt.title('Learning Curves for Different Sizes of Training Data')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

train_scores
```

The first trend we see is that there is a positive correlation between the amount of training data and the model's performance. Speicifcally, the model that trained with 100% of the training set had the lowest loss value over the 200 epochs. This improved performance is attributed to the broader range of examples and potentially more complex patterns that the model needs to learn. So, the model can capture the underlying distribution of the dataset more effectively, improving the ability to generalize to unseen data.

This experiment also showed the idea of diminishing returns with the increase in training data size. Specifically, significant improvements in the model's performance as well as loss reduction was seen when the size of the training data was doubled from 25% to 50%. However, there wasn't much improvement of performance when added more data beyond this point; this was clearly seen when comparing the model's performance on the 75% training data size to the full dataset. This suggests that the model is able to learn the critical features from a subset of the data, and additional data beyond a certain threshold contributes less to performance gains. This could mean that there exists an optimal amount of data for training, after which the value of adding more data goes down. 

The third trend we observed was the rapid initial decline in loss for models trained with smaller datasets. This steep intial loss curve implies that the models are quickly fitting the limited patterns present in smaller datasets. But, when it comes to the loss value after 200 epochs, these models had high loss values compared to the other higher % subsets. This shows that these models may not be generalizing well beyond their training data. 

Overall, the amount of training data plays a crucial role in the learning performance of neural network models. More data generally leads to better model performance, but the rate of improvement declines as the amount of data increases. In addition, using the whole subset of the training data usually leads to the best results. 

Let's delve deeper into the analysis of the general trends observed from the learning curve results, discussing the influence of training data amount and training duration.

1. **Influence of Training Data Amount**:
   One distinct trend observed is the relationship between the amount of training data and the model's ability to generalize, as indicated by the loss values. The model trained with 100% of the data demonstrates the lowest loss, whereas the model with 25% of the data exhibits the highest loss at convergence. This suggests that the richness and diversity of the full dataset provide a more comprehensive learning experience, allowing the model to capture the underlying distribution of the data more effectively. Conversely, the limited data scenarios are likely missing critical information, leading to poorer generalization and potentially higher susceptibility to noise and variance in the data.

2. **Diminishing Returns on Increased Data**:
   Another trend is the diminishing returns on model improvement with increasing amounts of training data. While we see a significant performance gain when comparing the 25% and 50% data scenarios, the difference between 75% and 100% is notably smaller. This pattern could indicate that, after a certain threshold, additional data points may not contribute novel information necessary for further reducing the loss. It is possible that the most critical patterns are already learned, and subsequent data points serve to reinforce these patterns rather than reveal new ones.

3. **Initial Learning Phase**:
   The initial epochs of training reveal a steep decline in loss, especially pronounced in models with less training data. This rapid improvement could be due to the model quickly fitting the limited patterns present in the smaller datasets. However, these gains plateau more swiftly than those seen in models with more data, suggesting that models with limited data are quickly finding a local optimum that does not generalize well, a hallmark behavior of overfitting.

4. **Training Duration and Overfitting**:
   Regarding the influence of training duration, all models appear to benefit from a longer training period initially, as evidenced by the continual decrease in loss. However, the curves start to plateau as they approach the minimum loss achievable given the complexity of the model and the information content of the data used. This indicates that extending the training duration beyond this point would not yield significant improvements and could potentially lead to overfitting, especially in smaller datasets where the model has fewer examples to learn from.

In summary, the amount of training data has a pronounced effect on the model's learning capability, with more data leading to better generalization. However, there is a point at which adding more data yields diminishing improvements. The initial learning phase is crucial, with faster improvements seen in smaller datasets, but these are limited by the potential for overfitting. Finally, training duration must be balanced to avoid overfitting, ensuring that the model is trained just enough to reach its performance potential without learning the noise within the training data.
